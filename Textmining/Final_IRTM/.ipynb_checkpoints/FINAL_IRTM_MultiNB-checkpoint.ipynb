```python
import sys
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['figure.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10
```

```python
nRowsRead = 1000 # 先看一部分就好

df = pd.read_csv('./mbti_datas/dataset1_clean_tokenize.csv', delimiter='\t').dropna()# , nrows = nRowsRead
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
df.reset_index(drop=True, inplace=True)
df.iloc[160, 0] #ESTP
df.iloc[100, 0] #INFP
```

#### 分割測試

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['clean_posts'], df['type'].values, test_size=0.1)
df_test = df.iloc[X_test.index].reset_index(drop = True)
df_train = df.iloc[X_train.index].reset_index(drop = True)
df_train
```

```python
priorC = df_train['type'].value_counts()/len(df_train) #事前機率P(C)
priorC
```

```python
type_list = ['INFJ', 'ENTP', 'INTP', 'INTJ', 'ENTJ', 'ENFJ', 'INFP', 'ENFP', 'ISFP', 'ISTP', 'ISFJ', 'ISTJ', 'ESTP', 'ESFP', 'ESTJ', 'ESFJ']
```

#### 所有字的字典

```python
Allwords = [] #一個 總體Dictionary
    
def buildDict(text): #只算有無出現, 弄成list
    for word in str(text).split(' '):
        #print(word)
        if word not in Allwords:
            Allwords.append(word)

df_train = df_train[df_train['clean_posts'].notnull()]
df_train.apply(lambda row: buildDict(row['clean_posts']), axis=1)
Allwords_dict = sorted(Allwords) #照字母排
AllwordSize = len(Allwords_dict) #總term數
AllwordSize
```

```python
eachCategorycnt_dict = {} #裡面存每個類別裡出現多少次每個字
wcount_dict = {} #計算每類別中'總字數'有多少
for i in range(len(type_list)): #照著順序走 先初始化
    dict_from_list = {k: 0 for k in Allwords_dict}
    eachCategorycnt_dict[type_list[i]]=dict_from_list
    wcount_dict[type_list[i]] = 0
    
    
for i in range(len(df_train)):  
    ctype = df_train.loc[i, 'type']
    for wd in df_train.loc[i, 'clean_posts'].split(' '): #對於每篇文章的每個term
        wcount_dict[ctype]+=1 #此類型總字數+1
        eachCategorycnt_dict[ctype][wd] += 1 #此類型的此字總數+1
    #print(wcount_dict)
    #print(eachCategorycnt_dict)
```

```python
import copy

params_dict = copy.deepcopy(eachCategorycnt_dict)
for category in range(len(type_list)):
    for k in params_dict[type_list[category]]:
        params_dict[type_list[category]][k] = (params_dict[type_list[category]][k]+1)\
        / (wcount_dict[type_list[category]]+AllwordSize)
#params_dict #計算每個字在每個類別中的smoothing後之分數 = P(t|c)
```

#### feature selection

```python
#先建立一個算所有terms在所有文章中出現次數加總
#先前已計算個類別中字典裡所有字出現次數在eachCategorycnt_dict
#其中可能有反指標
wordSum_dict = {k: 0 for k in Allwords_dict} #一個字典中有所有的訓練term以及總出現數
for wd in wordSum_dict:
    for i in range(len(type_list)):
        wordSum_dict[wd] += eachCategorycnt_dict[type_list[i]][wd] #每個class出現次數加總即為總出現次數
        
chi_score_dict = {k: 0 for k in Allwords_dict} #另創一個紀錄chi-square value 對於字典全部字
for wd in wordSum_dict:
    score = 0
    for i in range(len(type_list)):
        expected = wordSum_dict[wd] * priorC[type_list[i]]
        tmp = ((eachCategorycnt_dict[type_list[i]][wd]-expected)**2)/expected #期望：字出現頻率*此類佔多少
        score += tmp
    chi_score_dict[wd] = score
    
chi_score_dict = dict( sorted(chi_score_dict.items(), key=lambda x: x[1], reverse=True) ) #照大小排
chi_score_dict 
```

```python
chi_feature = []
for key in chi_score_dict.keys():
    chi_feature.append(key)
#imp_feature = chi_feature[:50000]
#chi_feature
```

```python
with open(f'./chiScore.txt', 'w') as wf:
    for item in chi_feature:
        wf.write("%s\n" % item)          
```

```python
def computeCat(featuresize=500):
    imp_feature = chi_feature[:featuresize]
    #計算每個被算對的類別，數目加起來除以總test比數則為accuracy
    typesCorrect = {}
    typesCnt = {}
    for i in range(len(type_list)):
        typesCorrect[type_list[i]] = 0
        typesCnt[type_list[i]] = 0

    #跑每一篇測試集 看有沒有對
    for i in range(len(df_test)):  
        text = df_test.loc[i, 'clean_posts'].split(' ')
        text = [x for x in text if x in imp_feature] #在重要文字中才納入
        correctType = df_test.loc[i, 'type']

        maxC = ''
        maxValue = -2e200
        for c in range(len(type_list)):
            tmpValue = 0
            tmpValue += math.log(priorC[type_list[c]])
            for wd in text:
                tmpValue += math.log(params_dict[type_list[c]][wd])
            if tmpValue > maxValue:
                maxValue = tmpValue
                maxC = type_list[c]
        if maxC == correctType:
            typesCorrect[correctType] += 1 #分對類別了
        typesCnt[correctType] += 1 #數一下這個類別在測試集有幾篇

    #print("\nsuccessfully retrieve:", typesCorrect)
    #print("\ntotal should retrieve:", typesCnt)

    totalcorrect = 0 #集合起來看共對了多少
    for cat, val in typesCorrect.items():
        totalcorrect += val

    totalcorrect /= len(df_test)
    print(f"{featuresize} feature terms, accuracy =", totalcorrect)
```

```python
fsize = [5, 10, 25, 50, 100, 250,  500, 1000, 2500, 6000, 12000, 30000, 60000]
print("total terms: ", AllwordSize)
for s in fsize:
    computeCat(s)
```

以上是用`chi_square Feature selection`選重要terms<br>
可以看到前15個重要變數都是<b>人格名稱</b>

```python
#type_list
def plot_Recall(Recall_cnt, All_cnt):    
    ALL = []
    Recall = []
    ratio = []
    for i in type_list:
        ALL.append(All_cnt[i])
        Recall.append(Recall_cnt[i])
        ratio.append(f'{Recall_cnt[i]}/{All_cnt[i]}')
        
    ax = plt.subplots(figsize=(7,7))
    ax = sns.barplot(x=np.arange(16), y=ALL, color='cornflowerblue')
    ax = sns.barplot(x=np.arange(16), y=Recall, color='cyan')

    # renaming the axes
    ax.set(ylabel="num of Documents")
    ax.set_xticklabels(type_list, rotation=45)
    addlabels(type_list, ALL, ratio)
    plt.title('Recall of each MBTI')
    plt.show()
    
def addlabels(x,y,r):
    for i in range(len(x)):
        plt.text(i, y[i]/2 ,r[i], ha = 'center', rotation=90, alpha=0.7)
```

此處僅是為了測試前500個features對各個category的取回程度如何

# 在此輸入文字與feature數即可算出分類


引用前處理。

```python
import re
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
word_list = ['is', 'am', 'are', 'was', 'were']
#stopwords from txt
my_file = open("stopwords.txt", "r")
stopwords = my_file.read()
stopwords = [stopwords.translate({ ord(c): None for c in '" ' }).split(",") ][0]#delete punctuation
```

```python
def tokenize(text): #順序很有差！以下都用空白取代
    text = re.sub(r'\|\|\|', r' ', text) #照數據格式切成一段一段
    text = re.sub(r'http\S+', r' ', text) #刪除網址
    text = re.sub('\S+com$', ' ', text) #刪除網站
    text = re.sub('\d+',' ', text) #刪除所有數字
    text = re.sub(r'[^\w\s]', ' ', text) #刪除所有標點符號 避免如ideas/thought -> ideasthought
    #print(text)
    #text = [i.translate({ ord(c): None for c in '"1234567890' }) for i in text] #delete punctuation
    #text = [i.translate({ ord(c): ' ' for c in "#$%&'()*+,-./:;!<=>?@[\]^_`{|~} " }) for i in text]
    text = re.sub("[^a-zA-Z]", " ", text) #只保留文字
    text = re.sub(' +', ' ', text) # Remove spaces > 1
    
    text = [i.strip() for i in text.split(' ')]  
    text = [i.lower() for i in text]
    text = [ps.stem(i) for i in text] #porter's algo
    text = ['be' if idx in word_list else idx for idx in text] #lemmatization
    text = [i for i in text if i not in stopwords] #stopwords delete
    text = list(filter(None, text)) #去除空字元
    stringtext = ' '.join([str(item) for item in text])
    #print(stringtext)
    return stringtext
```

```python
def computeLLH(text='', featuresize=500):
    
    imp_feature = chi_feature[:featuresize]
    text = str(text).split(' ')
    text = [x for x in text if x in imp_feature]
    
    tmp_cScores = []
    cScores = {}
    
    for c in type_list:
        tmpValue = 0
        tmpValue += math.log(priorC[c])
        for wd in text:
            tmpValue += math.log(params_dict[c][wd])
        #tmp_cScores.append(tmpValue) 為了要能展示％數
        tmp_cScores.append(math.exp(tmpValue)) #還原機率
    
    tmp_cScores = tmp_cScores/np.sum(tmp_cScores) #使機率加總為1
    
    for c in type_list:
        cScores[c] = tmp_cScores[0]
        tmp_cScores = tmp_cScores[1:]
    cScores = dict( sorted(cScores.items(), key=lambda x: x[1], reverse=True) ) #照大小排
    
    return cScores
```

```python
text = input("give me words: \n")
result = computeLLH(tokenize(text), 200)
for c, score in result.items():
    print(f"{c}:, {100*score}%")
```
