```python
#first train a model to adjust hyperparameter --- depth
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
for max_d in range(1,15): #試試看深度幾層好
    model = DecisionTreeClassifier(max_depth=max_d, random_state=42)
    model.fit(X_train, y_train)
    print('The Training Accuracy for max_depth {} is:'.format(max_d), model.score(X_train, y_train))
    print('The Validation Accuracy for max_depth {} is:'.format(max_d), model.score(X_test,y_test))
    print('')
print("it seems that set the depth to 2~4 is a good choice")
plt.figure(figsize =(80,20))
plot_tree(model, feature_names=X_train.columns, max_depth=2, filled=True)
```

```python
#kfold_model = model_selection.KFold(10, shuffle=True)
#print(model_selection.cross_val_score(model, X_test, y_test, cv=kfold_model))
from sklearn import model_selection
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix
model = DecisionTreeClassifier(max_depth=4, random_state=42)
cv = StratifiedKFold(n_splits=10,shuffle=False)

print("K-fold result：")
for train,test in cv.split(X,y):
    model.fit(X.iloc[train],y.iloc[train])
    predicted_y = model.predict(X.iloc[test])
    true_y = y.iloc[test]
    #print(predicted_y)
    #print(true_y)
    print(f"Accuracy：{metrics.accuracy_score(true_y, predicted_y)}, AUC：{metrics.roc_auc_score(true_y, model.predict_proba(X.iloc[test])[:,1])}, f1_score：{metrics.f1_score(true_y, predicted_y)}")
print(confusion_matrix(true_y, predicted_y)) #方便起見用最後一組來檢驗，幾乎都預測-1
print(metrics.classification_report(true_y, predicted_y))

print('almost all of the predictions are -1s, and the accuracy is still high')
confusion_matrix = metrics.confusion_matrix(true_y, predicted_y)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

### (d.) 
There are 104 1s and 1463 -1s, we can choose to use undersampling(1:3) to handle this imbalanced situation. However, if the performance isn't enhanced later, we might repeat undersampling for 5 (1463/312) times to fully utilize the dataset.

```python
failed_tests = np.array(df[df['Label'] == 1].index)
no_failed_size = 3*len(failed_tests)
normal_tests = np.array(df[df['Label'] == -1].index)

random_normal_indices = np.random.choice(normal_tests, size = no_failed_size, replace = True)
random_normal_indices = sorted(np.array(random_normal_indices)) #隨機取312個正確值

under_sample = sorted(np.concatenate([failed_tests, random_normal_indices]))
len(under_sample) #104+312=416
```

```python
under_sample_data = df.iloc[under_sample, :]
under_sample_data
```

```python
#(e.)
from sklearn.metrics import confusion_matrix

model = DecisionTreeClassifier(max_depth=4, random_state=42) #設定樹最大深度為4
cv = StratifiedKFold(n_splits=10,shuffle=False)

X = under_sample_data.iloc[:, :-1]#df.drop(['Label'], axis=1)
y = under_sample_data['Label']

AccuracyL = []
AUCL = []
F1L = []
print("K-fold result：")
for train,test in cv.split(X,y):
    model.fit(X.iloc[train],y.iloc[train])
    predicted_y = model.predict(X.iloc[test])
    true_y = y.iloc[test]
    AccuracyL.append(metrics.accuracy_score(true_y, predicted_y))
    AUCL.append(metrics.roc_auc_score(true_y, model.predict_proba(X.iloc[test])[:,1]))
    F1L.append(metrics.f1_score(true_y, predicted_y))
    print(f"Accuracy：{metrics.accuracy_score(true_y, predicted_y)}, AUC：{metrics.roc_auc_score(true_y, model.predict_proba(X.iloc[test])[:,1])}, f1-score：{metrics.f1_score(true_y, predicted_y)}")
print(confusion_matrix(true_y, predicted_y)) 
print(f"\nmean of Accuracy：{np.mean(AccuracyL)}, mean of AUC：{np.mean(AUCL)}, mean of F1-score：{np.mean(F1L)}")
#print(metrics.classification_report(true_y, predicted_y))
```
