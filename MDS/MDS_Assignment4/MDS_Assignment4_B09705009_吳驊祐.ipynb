```python
#接著移除變數間高相關
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.95) #相關係數超過0.9則刪除
X_train = X_train.drop(corr_features, axis=1)
X_test = X_test.drop(corr_features, axis=1)

#移除後剩下個386features
print(f'After removing {len(corr_features)} features, there are {X_train.shape[1]} features left.')
```

### Over sampling (SMOTE)

```python
def fit_result(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    predicted_y = model.predict(X_test)
    true_y = y_test

    print(f"Precision：{metrics.precision_score(true_y, predicted_y)}, \
            Recall：{metrics.recall_score(true_y, predicted_y)},\n \
       Accuracy：{metrics.accuracy_score(true_y, predicted_y)}, \
       AUC：{metrics.roc_auc_score(true_y, model.predict_proba(X_test)[:,1])}, \
       f1-score：{metrics.f1_score(true_y, predicted_y)}")
    cm = confusion_matrix(true_y, predicted_y)
    print(cm) 
    print(metrics.classification_report(true_y, predicted_y))
    
    #此處-1是先的label容易搞混，但我們就照平常的方法來不管此例是-1還是1
    fig, ax = plt.subplots(figsize=(5, 5))
    plot_confusion_matrix(model, X_test, y_test,  cmap=plt.cm.Blues, colorbar=False, ax=ax)
    plt.show()
    
    #已驗算過
    print('The TP rate is: {0:4.2}'.format(float(cm[0][0])/np.sum(cm[0])))
    print('The TN rate is: {0:4.2}'.format(float(cm[1][1])/np.sum(cm[1])))
    print('The FN rate is: {0:4.2}'.format(1- float(cm[0][0])/np.sum(cm[0]))) #FN = 1-TP
    print('The FP rate is: {0:4.2}'.format(1- float(cm[1][1])/np.sum(cm[1]))) #FP = 1-TN
```

### 基因演算法
基因演算法介紹，無程式 https://www.youtube.com/watch?v=Fdk7ZKJHFcI<br>

- 第一：初始化(initialization)階段，設定母體大小、交配率、突變率、演算法終止條件等。
- 第二：將欲求解的最佳化問題的解進行編碼(encoding)，並根據母體大小隨機產生多個染色體作為母體，形成初始的親代(parents)。
- 第三：開始展開迭代，染色體進行`交配與突變`從而產生子代(offspring)。
    - 也就是產生新的候選解，藉由增加解的多樣性來實踐對<b>解的探索</b>，同時也能使解逃脫出區域最佳解
        - 單點交配/多點交配/均勻交配
- 第四：每個染色體經解碼(decoding)後分別計算對應到的適應值，並選擇優良的保留<b>(解的開採)</b>。
    - 多使用Roulette wheel隨機射輪盤法，使得優良的有高機率被選中，同時差的也有機會，可避免過早成熟。
- 第五：若終止條件被滿足則演算法收斂。
    - 設定固定的迭代次數(number of iterations)
    - 設定當前最佳解的適應值在連續經歷幾個世代後依然不變的次數
    - 設定當前最佳解的適應值小於某個閥值(或改善幅度小於某個百分比)

```python
num_job=20 # number of jobs

p=[10,10,13,4,9,4,8,15,7,1,9,3,15,9,11,6,5,14,18,3]
d=[50,38,49,12,20,105,73,45,6,64,15,6,92,43,78,21,15,50,150,99]
w=[10,5,1,5,10,1,5,10,5,1,5,10,10,5,1,10,5,5,1,5]
```

```python
#計算weight of tardiness總和
def calculateVal(seq):
    now_time = 0
    total_val = 0
    for i in range(num_job):
        now_time = now_time + p[seq[i]]
        if d[seq[i]]<now_time: 
            total_val += (w[seq[i]]*now_time-d[seq[i]])
        #print(seq[i], total_val)
    return total_val
```

```python
#演算法
def genetic_alg(iternum, population_size, crossover_prob, mutate_prob):
    #隨機產生初始順序
    best_list=[] #最好的那個排列
    best_obj= 9999999 #最好排列的obj value
    population_list=[] #當前的母體有哪些, 全程每個iteration結束時都要維持數目為population_size!!!!!!!!!!!!!!!!!!
    for i in range(population_size):
        random_num=list(np.random.permutation(num_job)) #隨機產生0~19排列
        population_list.append(random_num) # add to the population_list
        if calculateVal(random_num) < best_obj: 
            best_list = random_num
            best_obj = calculateVal(random_num)
    print(f"initial best_obj:{best_obj}")
    recordnow = [best_obj]
    recordObj = [best_obj]
    for i in range(iternum): 
        #計算新增了多少child using cross
        parent_list=copy.deepcopy(population_list)
        child_list=[]
        #最後以上兩者要合併且選出30個回population_list，其餘丟棄
        
        S=list(np.random.permutation(population_size)) #隨機排列誰跟誰crossover
        for j in range(int(population_size/2)): #0~14
            p = np.random.rand()
            if crossover_prob>=p: #交配機率用在此
                k = np.random.randint(1,20) #single point crossover 此點後兩方交換，1~19是有用數字
                parent1 = parent_list[S[2*j]] #0,2,4,...,28
                parent2 = parent_list[S[2*j+1]] #1,3,5,...,29
                child1 = np.concatenate([parent1[:k],parent2[k:]])
                child2 = np.concatenate([parent2[:k],parent1[k:]])
                child_list.append(child1)
                child_list.append(child2)
        for j in range(population_size): #計算新增了多少child using mutation
            p = np.random.rand()
            if mutate_prob>=p: #突變機率用在此
                k = random.sample(list(np.arange(20)), 2) #取出0~19裡兩個不同的數字進行突變(交換兩者)
                child = parent_list[j]
                #print()
                #print('mutate child',child)
                child[k[0]], child[k[1]] = child[k[1]], child[k[0]]
                child_list.append(child)
                #print('mutate child',child)
                #print()
                
        #整合時間selection
        currentBest = 99999999
        total_chromosome=copy.deepcopy(parent_list)+copy.deepcopy(child_list)
        sizeofCandidate = len(total_chromosome)
        id_score = []
        population_list = []
        for j in range(sizeofCandidate):
            candi_score = calculateVal(total_chromosome[j])
            id_score.append([j, candi_score]) #建立一個順序分數的對應
            if candi_score < best_obj: #若此分數最高記得更新
                best_list = total_chromosome[j]
                best_obj = candi_score
            if candi_score < currentBest:
                currentBest = candi_score
        recordnow.append(currentBest)
                
        #現在要選populationsize的為新母體，前q名看排名，剩下的population從剩餘隨機選
        #沒機會實作roulette wheel
        q = 12
        id_score = sorted(id_score, key=lambda x:x[1], reverse=False) #依照分數排序，越低越好在前面
        for j in range(q):
            population_list.append(total_chromosome[j])
            
        randomList = np.random.choice(np.arange(q,sizeofCandidate), population_size-q, False)
        for j in randomList:
            population_list.append(total_chromosome[j])
        recordObj.append(best_obj)
    #print(recordObj)
    return best_list, best_obj, recordObj, recordnow
```

### 禁忌搜尋法
在禁忌搜尋法中，鄰近解的生成方式與禁忌名單大小(Tabu list size)對於最佳解的搜尋有顯著的影響，此二者決定了可行解探索的方式與程度。<br>
⚫常見的鄰近解的生成方式包含了鄰近互換(neighborhood swap)以及 兩兩互換(random swap)等，而這通常會需依據問題的特性來設定。
- 禁忌名單大小的設定
   - 若名單大小設定較大，則將使得鄰近解強調比對過去所搜尋過的路徑，不走回頭路，<br>
       使其探索的程度相對大於開採(解搜尋會收斂往特定的方向前行) ，降低局部的搜尋
   - 相對地，若禁忌名單大小設定較小時，將經常遺忘過去所走過的路徑，加強鄰近區域開採的能力，而減少探索未知區域。
   - 因此，在禁忌搜尋法中適當的選擇禁忌名單大小(超參數)會是取得探索與開採權衡的關鍵

```python
def taboo_alg(iternum, tabu_size):
    #產生一組初始解 隨機產生0~19排列
    now_sequence=list(np.random.permutation(num_job)) #每次iteration都會改變當下sequence
    
    best_list = now_sequence #永遠紀錄最好的那個排列
    best_obj = calculateVal(now_sequence) #最好排列的obj value
    print(f"initial best_obj:{best_obj}")
    
    recordnow = [best_obj] #紀錄每次iteration中最好的obj value
    recordBest = [best_obj] #記錄到每個iteration，得到過最好的obj value
    taboos = [] #每次加入時append在最後面，並把最前面的（也就是裡面最老舊的)pop掉 
    for i in range(tabu_size):
        taboos.append((0,0)) #先以無用資訊填滿，用來統一之後操作
    
    for i in range(iternum):       
        tmpBestValue = 9999999
        tmpBestSwitch = 0 #紀錄這次要讓哪兩個鄰居換，注意位置和實際值要搞清楚(這邊紀錄位置就好)
        for j in range(num_job-1): #讓j和j+1互換 (neighborhood swap)，選出最好的
            tmpSequence = now_sequence
            tmpSequence[j], tmpSequence[j+1] = tmpSequence[j+1], tmpSequence[j]
            val = calculateVal(tmpSequence)
            #走每次iteration中最好的，且不違反禁忌的那步 如(1,5)(5,1)都不能
            if val < tmpBestValue and (tmpSequence[j], tmpSequence[j+1]) not in taboos \
                    and (tmpSequence[j+1], tmpSequence[j]) not in taboos: 
                tmpBestValue = val
                tmpBestSwitch = j
        #選擇完要換哪個了，更新現在走到哪，更新禁忌
        taboos.pop(0)
        taboos.append((now_sequence[tmpBestSwitch], now_sequence[tmpBestSwitch+1]))
        now_sequence[tmpBestSwitch], now_sequence[tmpBestSwitch+1] = now_sequence[tmpBestSwitch+1], now_sequence[tmpBestSwitch]

        #看這次的最好有沒有比之前更好
        if tmpBestValue < best_obj:
            best_obj = tmpBestValue
            best_list = now_sequence
            
        recordnow.append(tmpBestValue)
        recordBest.append(best_obj)
    return best_list, best_obj, recordBest, recordnow
```

```python
iternum= 1000 
tabu_size = 20

bList, bObj, recordBest, recordnow = taboo_alg(iternum, tabu_size)
```

```python
print("best Object value(min weighted tardy):", bObj)
print("best Sequence:", bList)

plt.style.use('seaborn-darkgrid')
plt.title("Sensitive analysis", loc='left', fontsize=12, fontweight=0, color='darkorange')
plt.xlabel("generation")
plt.ylabel("Obj value")

p1 = plt.plot(recordBest, c='navy', marker='v',label='alwaysbestObj' )
p2 = plt.plot(recordnow, c='cyan',label='currentbestObj', linewidth=0.3 )

plt.legend(loc="upper left", facecolor='blue')
plt.show()
```
